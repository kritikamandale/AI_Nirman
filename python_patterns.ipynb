```python
#1 pattern
n = int(input("enter the value = "))
for i in range(1, n):
    for j in range(1 , n):
        print("*", end = " ")
    print("\n")
```

```python
#2 pattern
n = int(input("enter the value = "))
for i in range(1, n):
    for j in range(1,i):
        print("*", end = " ")
    print("\n")
```

```python
#3 pattern
n = int(input("enter the value = "))
for i in range(1, n):
    for j in range(1,i):
        print(j, end = " ")
    print("\n")
```

```python
# 4 pattern
n = int(input("enter the value = "))
for i in range(1, n+1):
    for j in range(i,n+1):
        print("*",end = "")
    print()
```

```python
#5 pattern
n = int(input("enter the value = "))
num = 1
for i in range(1, n+1):
    for j in range(i):
        print(num,end = " ")
        num += 1
    print()
```

```python
#8 pattern
n = int (input( "enter the number : "))
for i in range (1, n+1):
  for j in range(1, n+1):
    if ( i == 1 or i == n or j == 1):
      print("C", end = " ")
    else:
      print(" ", end = " ")

  print()
```

```python
#9 pattern
n = int(input("enter the value = "))
for i in range(1, n+1):
  print(" "*(n-i), end = " ")
  for j in range(1, i+1):
        print( chr(64 + j), end = " ")
  print("\n")
```
